```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv("DADOS/trees.csv")
base.head()
```

```python
plt.scatter(x=base.Girth, y=base.Volume, marker="*")
plt.title("ARVORES")
plt.ylabel("CIRCUFERENCIA")
plt.xlabel("VOLUME")
```

```python
plt.scatter(x=base.Girth,y=base.Volume,marker="*")
```

```python
plt.plot(base.Girth,base.Volume)
plt.title("ARVORES")
plt.ylabel("CIRCUFERENCIA")
plt.xlabel("VOLUME")

#GRAFICO DE LINHA DE TENDENCIA
```

```python
sns.regplot(x=base.Girth,y=base.Volume, data=base, x_jitter = 0.3, fit_reg = False)
```

```python
sns.regplot(x=base.Girth,y=base.Volume, data=base, x_jitter = 0.3, fit_reg=True)
```
