```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv("DADOS/trees.csv")
base.head()
```

```python
plt.boxplot(base.Volume, vert = False,showfliers = False, notch = False,patch_artist =False)
plt.title("ARVORES")
plt.xlabel("VOLUME")
```

```python
plt.boxplot(base.Volume, vert=True, showfliers = True, notch = True, patch_artist=True)
plt.title("ARVORE")
plt.xlabel("VOLUME")
```

```python
plt.boxplot(base.Volume, vert = False, showfliers= False, notch=True, patch_artist=True)
plt.title("ARVORES")
plt.xlabel("VOLUME")
```

```python
plt.boxplot(base.Volume)
plt.title("ARVORES")
plt.xlabel("VOLUME")
```

```python
#GERANDO 3 BOXPLOT COM INFORMAOES DIFERENTE

plt.boxplot(base.Volume, vert=False)
plt.boxplot(base.Girth, vert= False)
plt.boxplot(base.Height, vert=False)
plt.title("ARVORE")
plt.xlabel("DADOS")
```

```python
plt.boxplot(base.Volume,  vert=True)
plt.boxplot(base.Girth,vert=True)
plt.boxplot(base.Height, vert=True)
plt.title("ARVORES")
plt.xlabel("DADOS")
```
